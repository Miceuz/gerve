# Pradiniai duomenys

```python
radius = 1.86 #m
mass = 50 #kg

maxToruqeAvailable = 60 #Nm

displacement = pi*radius #m pusę apsisukimo
print "kameros nueitas kelias: ", displacement

acceleration = maxToruqeAvailable / (2 * radius * mass) #0.005 #m/s**2
print "maksimalus galimas pagreitis pagal turimą sukamąjį momentą ", maxToruqeAvailable, "Nm:", acceleration, "m/s**2"

midtime = int(sqrt(displacement / acceleration)) #s išsiskaičiuojam, kada jau pradėt stabdyt
print "midtime:", sqrt(displacement / acceleration), midtime
```

# Prasimuliuojam pabraižom

```python
speed = np.zeros(midtime*2+1)
position = np.zeros(midtime*2+1)
for t in range(0, midtime*2+1):
    if t <= midtime:
        speed[t] = acceleration * t
    else:
        speed[t] = speed[midtime] - acceleration * (t - midtime)
    position[t] = sum(speed[:t+1]) #pozicija yra greičio integralas
    #print t, speed[:t], sum(speed[:t]), position[t]
print "didžiausias greitis",speed[midtime], "m/s"
plot(speed)
title("Gerves greitis besisukant puse lanko")
xlabel("laikas, s")
ylabel("greitis, m/s")
figure()

plot(position)
title("Kameros nueitas kelias")
xlabel("laikas, s")
ylabel("nueitas kelias, m")

print(max(position))
```
